```python
# 1.二叉树的深度
# 输入一棵二叉树，求该树的深度。从根结点到叶结点依次经过的
# 结点（含根、叶结点）形成树的一条路径，最长路径的长度为树的深度，根节点的深度视为 1 。
'''
思考：分别递归求左右子树的深度即可
'''
class TreeNode:
    def __init__(self, x):
        self.val = x
        self.left = None
        self.right = None
        
class Solution:
    def TreeDepth(self , tree: TreeNode) -> int:
        # 相当于 if root = None:
        if not tree:
            return 0
        if not tree.left and not tree.right:
            return 1
        return 1 + max(self.TreeDepth(tree.left), self.TreeDepth(tree.right))

tree = {1,2,3,4,5,2,6, 2,2 ,7}

# obj.TreeDepth(tree)
# 这样传参数肯定是错的，题目中定义的函数传入的参数是treenode，不是整棵树，所以需要自己构建一棵树，传入根节点
obj = Solution()
node1 = TreeNode(1)
node2 = TreeNode(2)
node3 = TreeNode(3)

node1.left = node2
node1.right = node3

obj.TreeDepth(node1)
```

```python
# 2. 按照之字形顺序打印二叉树
# 给定一个二叉树，返回该二叉树的之字形层序遍历，（第一层从左向右，下一层从右向左，一直这样交替）
'''
思考：广度优先算法，每一层在一行输出，只需要将奇数行的result反转即可
    使用的是递归的方法，递归一般会比迭代好用

'''

class Solution:
    def Print(self , pRoot: TreeNode) -> list[list[int]]:
        if not pRoot:
            return []
        res = []
        self.level(pRoot, 0, res)
        # 反转基数行,level表示每一层的
        for level in range(1, len(res), 2):
            res[level] = res[level][::-1]
        return res
    def level(self, root, level, res):
        if not root:
            return
        # res是一个列表形式，列表中每一个元素是一个矩阵，初始化res过程稍加注意
        if level == len(res):
            res.append([])
        res[level].append(root.val)
        # 画个图就ok
        if root.left:
            self.level(root.left, level + 1, res)
        if root.right:
            self.level(root.right, level + 1, res)

obj = Solution()
node1 = TreeNode(1)
node2 = TreeNode(2)
node3 = TreeNode(3)

node1.left = node2
node1.right = node3

obj.Print(node1)
```

```python
# 3. 二叉搜索树的第k个节点
# 给定一棵结点数为n 二叉搜索树，请找出其中的第 k 小的TreeNode结点值。时间复杂度空间复杂度都为O-n
'''
1.返回第k小的节点值即可
2.不能查找的情况，如二叉树为空，则返回-1，或者k大于n等等，也返回-1
3.保证n个节点的值不一样

思考：节点的有序性帮我们确定了搜索的方向，
如果是找第k大的，可以使用中序遍历，按照从小到大的顺序到第k个停止即可
另：中序遍历步骤：
递归左子树-存储跟节点-递归右子树
then：本题的思路be like：
递归左子树，判断是否返回节点，如果有，停止递归返回所要的节点
判断当前根节点是否第k个遍历到的值，
递归右子树，判断有无返回节点
(一定注意，结构都是一样的！)
'''
class treeNode:
    def __init__(self, x):
        self.val = x
        self.left = None
        self.right = None
        
class Solution(treeNode):
    def __init__(self):
        # 记录返回的节点 
        self.res = None
        # 记录中序遍历了多少个
        self.count = 0
    def midOrder(self, root, k):
        # empty or over k, return
        if not root or self.count > k:
            return
        self.midOrder(root.left, k)
        self.count += 1
        # 只记录第k个
        if self.count == k:
            self.res = root
        self.midOrder(root.right, k)
        
    def KthNode(self, proot: treeNode, k: int) -> int:
        self.midOrder(proot, k)
        if self.res:
            return self.res.val
        # empty or not found
        else:
            return -1
        
tree = [5,3,7]
obj1 =  treeNode(5)
obj11 = treeNode(3)
obj22 = treeNode(7)

obj1.left = obj11 
obj1.right = obj22
obj = Solution()
print(obj.KthNode(obj1, 2))
```

```python
class TreeNode:
    def __init__(self, x):
        self.val = x
        self.left = None
        self.right = None
        
class Solution:
    def reConstructBinaryTree(self , pre: list[int], vin: list[int]) -> TreeNode:
        if not pre:
            return None
        #  root node
        root = TreeNode(pre[0])
        # 根节点在中序遍历中的位置索引，因为是list数据结构，所以可以用index
        tmp = vin.index(pre[0])
        # 递归，构造左子树
        root.left = self.reConstructBinaryTree(pre[1:tmp+1], vin[:tmp])
        # 递归，构造右子树
        root.right = self.reConstructBinaryTree(pre[tmp+1:], vin[tmp+1:])
        return root
    # 层次遍历将二叉树打印，前面做过的之字形
    def Print_level(self, root: TreeNode) -> list[list[int]]:
        res = []
        def helper(root, depth):
            if not root: res[depth].append(0)
            if len(res) == depth: res.append([]) # start the current depth
            res[depth].append(root.val) # fulfil the current depth
            if  root.left: helper(root.left, depth + 1) # process child nodes for the next depth
            if  root.right: helper(root.right, depth + 1)
        helper(root, 0)
        return res
            
        

obj = Solution()
pre = [1,2,4,7,3,5,6,8]
vin = [4,7,2,1,5,3,8,6]
root = obj.reConstructBinaryTree(pre, vin)
res = obj.Print_level(root)
print(res)
# 有一个小问题，上面重建的二叉树貌似没有给空的叶节点变成none，所以没法输出空
```

```python
class TreeNode:
    def __init__(self, x):
        self.val = x
        self.left = None
        self.right = None
class Solution:
    def Mirror(self , pRoot: TreeNode) -> TreeNode:
        if not pRoot:
            return pRoot
        # swap
        pRoot.left, pRoot.right = pRoot.right, pRoot.left
        # 递归
        self.Mirror(pRoot.left)
        self.Mirror(pRoot.right)
        return pRoot
    
```

```python
# 7. 从上到下打印二叉树
# 不分行从上往下打印出二叉树的每个节点，同层节点从左至右打印。之前做过分行的，通过depth实现顺序不乱，这样不分行会难一些
# @下面有分层打印，思路是一样的，就是多了分层操作
'''
思考：层次遍历，一般会用队列来辅助，队列是一种先进先出的数据结构。按照性质，从左到右访问一行节点，访问的时候依次把子节点加入队列，
子节点是从左到右，并且排在本节点后面，队列出现的顺序也是从左到右，符合层次遍历特点。
steps：
1 首先判断二叉树是否为空，
2 建立辅助队列，根节点进入队列，
3 每次遍历队首节定，如果有子节点，加入队列排队等待访问

画个图比较明
'''
import queue
class TreeNode:
    def __init__(self, x):
        self.val = x
        self.left = None
        self.right = None
class Solution:
    def PrintFromTopToBottom(self , root: TreeNode) -> List[int]:
        res = []
        if not root:
            return res
        # queue stroed, level order
        q = queue.Queue()
        q.put(root)
        while not q.empty():
            # 把当前节点拿走
            cur = q.get()
            res.append(cur.val)
            # 若左右孩子存在，存入左右孩子作为下一层次
            if cur.left:
                q.put(cur.left)
            if cur.right:
                q.put(cur.right)
        return res
```

```python
# 8. 二叉搜索树的后序遍历序列
# 输入一个整数数组，判断该数组是不是某二叉搜索树的后序遍历的结果。
# 如果是则返回 true ,否则返回 false 。假设输入的数组的任意两个数字都互不相同。
'''

'''

class Solution:
    def VerifySquenceOfBST(self , sequence: list[int]) -> bool:
        if not sequence:
            return False
        # digui
        return self.check(sequence, 0, len(sequence)-1)
    
    # left, right is sequence.index
    def check(self, sequence, left, right):
        if left >= right:
            return True
        # 找到第一个大于根节点的元素
        mid = left
        root = sequence[right]
        while sequence[mid] < root:
            mid += 1
            tmp = mid
        # 到此已经获取了mid的值可以左右子树递归判断了！
        # 保障postorder[mid]后的数大于root，否则返回false，
        while tmp < right:
            tmp += 1 
            if sequence[tmp] < root: 
                return False 
        # 递归左右子树进行判断 
        return self.check(sequence, left, mid - 1) and self.check(sequence, mid, right - 1)
```

```python
'''
递归三段式：
1 终止条件：到达叶节点，检查有无子节点，在判断路径值是否相等
2 返回值：子问题是否有符合新目标值的路径上返回
3 本级任务：检查是否到叶节点，没有就递归进入叶节点，更新sum减去该层val

clean and tidy code guaranteed！
'''
class Solution:
    def hasPathSum(self , root: TreeNode, sum: int) -> bool:
        if not root:
            return False
        if not root.left and root.right and sum - root.val == 0:
            return True
        # 递归进入子节点
        return self.hasPathSum(root.left, sum - root.val) or self.hasPathSum(self.right, sum - root.val)
    
```

```python
# 衍生问题：输入一颗二叉树的根节点root和一个整数expectNumber，找出二叉树中结点值的和为expectNumber的所有路径。
'''
思路：
深度优先和广度优先都可以做！
1）深度优先递归：和上面的做法一样！实现细节上多了需要维护两个向量，ret和path，递归函数内部需要更新path和expect number的值
    判断是否为叶结点，判断是否将true path加入ret
2）广度优先递归：使用队列，先入先出，层次搜索，维护逐层信息，遍历到叶节点，进行目标值匹配

'''
class TreeNode:
    def __init__(self, x):
        self.val = x
        self.left = None
        self.right = None
# method1
class Solution:
    def FindPath(self , root: TreeNode, target: int) -> List[List[int]]:
        ret, path = [], []
        def dfs(root, TreeNode, target: int):
            # empty
            if not root:
                return
            path.append(root.val)
            target -= root.val
            # 
            if not root.left and root.right and target == 0:
                ret.append(path[:])
            dfs(root.left, target)
            dfs(root.right, target)
            path.pop()
        dfs(root, target)
        return ret
# method2
    def FindPath2(self, root: TreeNode, target: int):
        if not root:
            return []
        # 将所需要的节点，路径和节点值组成一个列表
        s = [(root,[root.val], root.val)]
        ret = []
        while s:
            node, val, sum = s.pop()
            # 到叶节点，且路径和一致
            if node and not node.left and not node.right and sum == target:
                ret.append(val)
            # left path and right path
            if node.left:
                s.append(node.left, val + [node.left.val], sum + node.left.val)
            if node.right:
                s.append(node.right, val + [node.right.val], sum + node.right.val)
        return ret       
        
```

```python
class Solution:
    head = None
    pre = None
    def Convert(self , pRootOfTree ):
        if not pRootOfTree:
            # 中序递归，叶子为空则返回
            return None
        # 首先递归到左边最小值
        self.Convert(pRootOfTree.left)
        # 找到最小值，初始化head和pre
        if not self.pre:
            self.head = pRootOfTree
            self.pre = pRootOfTree
        # 当前节点与上一节点建立连接，将pre设置为当前值
        else:
            self.pre.right = pRootOfTree
            pRootOfTree.left = self.pre
            # 以上两步完成双向链接后，将当前设置为pre
            self.pre = pRootOfTree
        self.Convert(pRootOfTree.right)
        return self.head
# 这样 时间复杂度和空间复杂度 都是On
```

```python
# 11. 判断是否是平衡二叉树
# 平衡二叉树（Balanced Binary Tree），具有以下性质：它是一棵空树或它的左右两个子树的高度差的绝对值不超过1，
# 并且左右两个子树都是一棵平衡二叉树。
'''
思路：
1）自顶向下，递归，需要看左右子树是否是，如果能求出每个节点为根的树的高度，根据左右子树高度差判断是否满足
2）自底向上，现求出每个节点为根的树的高度，然后判断，或在求高度的同时判断，//利用后序遍历，递归到叶节点，回溯判断

建议看方法一就ok
'''
class TreeNode:
    def __init__(self, x):
        self.val = x
        self.left = None
        self.right = None
class solution:
    # method 1 自顶向下递归
    
    def IsBalanced_Solution(self, root: TreeNode) -> bool:
        if not root: return True
        # 高度差不能大于1，并且左右子树也是同样高度差要小于等于1
        return abs(self.depth(root.left) - self.depth(root.right)) <= 1 and \
            self.IsBalanced_Solution(root.left) and self.IsBalanced_Solution(root.right)
 
    def depth(self, root):
        if not root: return 0
        # 左右子树中高度较大的作为当前节点高度
        return max(self.depth(root.left), self.depth(root.right)) + 1
    
    # method 2 自底向上求高度,
    def IsBalanced_Solution(self, pRoot):
        # write code here
        return self.depth(pRoot)!=-1
    def depth(self,pRoot):
        if pRoot is None:
            return 0
        left = self.depth(pRoot.left)
        if left == -1:
            return -1
        right = self.depth(pRoot.right)
        if right == -1:
            return -1
        if (left-right) <-1 or (left-right)>1:
            return -1
        else:
            return 1 + left if left>right else 1+right
        
```

```python
# 13. 对称的二叉树
# 给定一棵二叉树，判断其是否是自身的镜像（即：是否对称）
'''
思考：递归，前序遍历
终止条件：进入子问题的两个节点都为空，说明到叶节点，且同步，返回true，有一个为空或者不等，返回false
返回值：每一级将子问题是否匹配结果向上传
本级任务：根左右 走左边，根右左 走右边， 一起进入子问题，需要两边都匹配
'''
```

```python
class TreeNode:
    def __init__(self, x):
        self.val = x
        self.left = None
        self.right = None
class Solution:
    def recursion(self, root1: TreeNode, root2: TreeNode):
        if not root1 and root2:
            return True
        # 只有一个为空或者不等
        if not root1 or not root2 or root1.val != root2.val:
            return False
        return self.recursion(root1.left, root2.right) and  self.recursion(root1.right, root2.left)
    def isSymmetrical(self , pRoot: TreeNode) -> bool:
        return self.recursion(pRoot, pRoot)
```

```python
# 14. 把二叉树打印成多行
# 给定一个节点数为 n 二叉树，要求从上到下按层打印二叉树的 val 值，同一层结点从左至右输出，
# 每一层输出一行，将输出的结果存放到一个二维数组中返回。
'''
思考：和之前的之字形打印有点像，这个难点在于在哪分层
有level 函数可以判断层数
用队列，每一层push进队列，具体做法参考第七题
多了一步：没进入一层记录队列长度，即本层节点数，访问相应节点放在一个数组，//key：子节点加入队列继续排队//
'''
class TreeNode:
    def __init__(self, x):
        self.val = x
        self.left = None
        self.right = None
class Solution:
    def Print(self , pRoot: TreeNode) -> list[list[int]]:
        res = []
        if pRoot is None:
            return res
        # 队列存储，进行层次遍历
        q = [pRoot]
        while q:
            # 记录某一行
            row = []
            n = len(q)
            for i in range(n):
                # 取出队列的队首
                node = q.pop(0)
                row.append(node.val)
                # 若存在左右孩子，加入队尾
                if node.left:
                    q.append(node.left)
                if node.right:
                    q.append(node.right)
            res.append(row)
        return res
            
            
```

```python
# 15. 序列化二叉树
# 请实现两个函数，分别用来序列化和反序列化二叉树，
# 不对序列化之后的字符串进行约束，但要求能够根据序列化之后的字符串重新构造出一棵与原二叉树相同的树。

# 可以考虑使用前序遍历解决！
class TreeNode:
    def __init__(self, x):
        self.val = x
        self.left = None
        self.right = None
class Solution:
    def __init__(self):
        self.index = 0
        self.s = ''
    def SerializeFunction(self, root):
        
        if not root:
            self.s += '#'
            return 
        # 根节点
        self.s += str(root.val) + '!'
        # 左子树
        self.SerializeFunction(root.left)
        # 右子树
        self.SerializeFunction(root.right)
        
    def Serialize(self, root):
        if not root:
            return '#'
        self.s = ''
        self.SerializeFunction(root)
        return self.s
    
    def DeserializeFunction(self, s: str):
        # 递归：到叶节点构建完毕，返回构建父节点
        # 为空
        if self.index >= len(s) or s[self.index]:
            self.index += 1
            return None
        # 数字转换
        val = 0
        # 遇到分隔符or结尾
        while s[self.index] != '!' and self.index != len(s):
            val = val*10 + (int)(s[self.index])
            self.index += 1
        root = TreeNode(val)
        # 序列到底，构建完成
        if self.index == len(s):
            return root
        else:
            self.index += 1
        # 反序列化与序列化一致都是前序
        root.left = self.DeserializeFunction(s)
        root.right = self.DeserializeFunction(s)
        return root
    
    def Deserialize(self, s: str):
        if s == '#':
            return None
        return self.DeserializeFunction(s)
```

```python
# 16. 二叉树中和为某一值的路径
# 给定一个二叉树root和一个整数值 sum ，求该树有多少路径的的节点值之和等于 sum 。
# 要求：不需要从根节点开始，叶节点结束，
'''
思考：所有节点都有可能是开始点，
1）递归（两次遍历）。查找路径需要向下遍历，继续前序遍历，遍历一个点sum减少，到最后一个节点刚好等于剩下sum，bingo
2）哈希表（一次遍历）。和之前用哈希表做路径的题思想一样
'''
# method 1
class TreeNode:
    def __init__(self, x):
        self.val = x
        self.left = None
        self.right = None
class Solution:
    def __init__():
        self.res = 0
    
    def dfs(self , root: TreeNode, sum: int) -> int:
        if root is None:
            return
        # 符合目标值
        if sum == root.val:
            self.res += 1
        # 进入子节点继续找
        self.dfs(root.left, sum - root.val)
        self.dfs(root.right, sum - root.val)
    def FindPath(self , root: TreeNode, sum: int) -> int:
        # 为空
        if root is None:
            return self.res
        # 查询以某节点为根的路径
        self.dfs(root, sum)
        # 以其子节点为新根
        self.FindPath(root.left, sum)
        self.FindPath（root.right, sum)
        return self.res

# method 2
class Solution:
    def __init__(self):
        #记录路径和及条数
        self.mp = dict()
        
    #last为到上一层为止的累加和
    def dfs(self, root: TreeNode, sum: int, last: int) -> int:
        #空结点直接返回
        if root is None: 
            return 0
        res = 0
        #到目前结点为止的累加和
        temp = root.val + last; 
        #如果该累加和减去sum在哈希表中出现过，相当于减去前面的分支
        if (temp - sum) in self.mp:
            #加上有的路径数
            res += self.mp[temp - sum] 
        #增加该次路径和
        if temp in self.mp:
            self.mp[temp] += 1
        else:
            self.mp[temp] = 1 
        #进入子结点
        res += self.dfs(root.left, sum, temp) 
        res += self.dfs(root.right, sum, temp) 
        #回退该路径和，因为别的树枝不需要这边存的路径和
        self.mp[temp] -= 1 
        return res
    
    def FindPath(self , root: TreeNode, sum: int) -> int:
        #路径和为0的有1条
        self.mp[0] = 1
        return self.dfs(root, sum, 0)
```

```python
# 17. 二叉树中找到两个节点的最近公共祖先
# 给定一棵二叉树(保证非空)以及这棵树上的两个节点对应的val值 o1 和 o2，请找到 o1 和 o2 的最近公共祖先节点。
'''
思路：
1）对root进行两次dfs，找到到o1o2的路径，路径最后一个公共节点就是最近公共祖先
2）递归。
'''
class TreeNode:
    def __init__(self, x):
        self.val = x
        self.left = None
        self.right = None
        
# method 1
class Solution:
    
    flag = False
    def dfs(self, root: TreeNode, path: list[int], o: int):
        if self.flag or not root:
            return
        path.append(root.val)
        # 因为每个节点值都不同，所以可以用val判断
        if root.val == o:
            self.flag = True
            return
        # dfs遍历寻找
        self.dfs(root.left, path, o)
        self.dfs(root.right, path, o)
        # 找到
        if self.flag:
            return
        # 所在子树没有，回溯
        path.pop()
    def lowestCommonAncestor(self , root: TreeNode, o1: int, o2: int) -> int:
        path1, path2 = [], []
        # 求根节点到两个节点的路径
        self.dfs(root, path1, o1)
        # 重置flag，查找下一个
        self.flag = False
        self.dfs(root, path2, o2)
        i = 0
        res = None
        # 从根开始，比较两个路径，找到第一个不同的点
        while i < len(path1) and i < len(path2):
            if path1[i] == path2[i]:
                # 最后一个相同节点是最近公共
                res = path1[i]
                i += 1
            else:
                break
        return res
    
```

```python
# 18. 二叉搜索树的最近公共祖先
# 给定一个二叉搜索树, 找到该树中两个指定节点的最近公共祖先。
# 多了一个搜索树的条件
'''
思路：搜索路径比较法，和上面一题一样
'''
class Solution:
    def getPath(self, root: TreeNode, target: int) -> List[int]:
        path = []
        node = root
        # 节点值都不同，可以直接用值比较
        while node.val != target: 
            path.append(node.val)
            # 小的在左子树
            if target < node.val: 
                node = node.left
            # 大的在右子树
            else: 
                node = node.right
        path.append(node.val)
        return path
    
    def lowestCommonAncestor(self , root: TreeNode, p: int, q: int) -> int:
        # 求根节点到两个节点的路径
        path_p = self.getPath(root, p) 
        path_q = self.getPath(root, q)
        # 比较两个路径，找到第一个不同的点
        i = 0
        while i < len(path_p) and i < len(path_q):
            if path_p[i] == path_q[i]:
                # 最后一个相同的节点就是最近公共祖先
                res = path_p[i]; 
                i += 1
            else:
                break
        return res
```
