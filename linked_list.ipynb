```python
# 2. 反转链表- easy
# description：定义一个函数，输入一个链表的头节点，反转该链表并输出反转后链表的头节点。
# example：输入: 1->2->3->4->5->NULL；输出: 5->4->3->2->1->NULL
"""
  @思考：重新定义新链表，实现链表元素的反转浪费内存，so改变链表next指针的指向即可。逐个进行即可，本科作业写过。
  两种方法：迭代法，也就是一个个来，递归法：稍微复杂
"""

# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, x):
#         self.val = x
#         self.next = None

# 迭代法
class Solution:
    def reverseList(self, head: ListNode) -> ListNode:
        cur = head
        pre = None
        while(cur!=None):
            # save cur.next since cur.next changes in the next step
            temp = cur.next
            # reverse operation
            cur.next = pre
            # update pre cur, make temp be cur (cur is changed)
            pre = cur
            cur = temp
            # return the head, now the head is pre, cur is None
        return pre
```

```python
# 3. 合并两个排序的链表
# description：输入两个递增的链表，单个链表的长度为n，合并这两个链表并使新链表中的节点仍然是递增排序的。
class Solution:
    def Merge(self , pHead1: ListNode, pHead2: ListNode) -> ListNode:
        if not pHead1:
            return pHead2
        if not pHead2:
            return pHead1
        result = ListNode(-1)
        cur = result
        while pHead1 and pHead2:
            if pHead1.val <= pHead2.val:
                cur.next = pHead1
                pHead1 = pHead1.next
            else:
                cur.next = pHead2
                pHead2 = pHead2.next
            # 指针右移，判断下一轮
            cur = cur.next
        # 将剩下的较大的节点拼接
        cur.next = pHead1 if pHead1 else pHead2
        # 返回头节点
        return result.next 
```

```python
# 4. 两个链表的第一个公共节点
# description：输入两个无环的单向链表，找出它们的第一个公共结点，如果没有公共节点则返回空。
'''
思考：两个指针，p1从N1头节点开始遍历，p2从N2头节点开始遍历，p1遍历完N1，接着从N2头开始遍历，P2从N1
    由于路程相同，所以一定会相遇，相遇的点就是公共节点，如图可以更好理解why

'''
class Solution:
    def FindFirstCommonNode(self , pHead1 , pHead2 ):
        a = pHead1
        b = pHead2
        # 两者相同则是第一个公共节点
        while a!=b:
            a = a.next if a else pHead2
            b = b.next if b else pHead1
        return a
```

```python
# 5. 链表中环的入口节点
# description：给一个长度为n链表，若其中包含环，请找出该链表的环的入口结点，否则，返回null。
'''
思考：哈希表思路：遍历会出现循环，想获得环入口节点，可以使用hash保存出现节点，重复遍历过程，第一次重复的节点就是环入口节点
     双指针思路：快慢指针相遇点，得到相关的关系,但需要考虑，第一次相遇的时候快指针可能已经转了n圈了（n>=1）
'''
# method1：哈希表， method2：快慢指针
class Solution:
    def EntryNodeOfLoop1(self, pHead):
        tags = set()
        while pHead:
            # 判断当前节点是否被访问过
            if pHead in tags:
                return pHead
            tags.add(pHead)
            pHead = pHead.next
        return None
    
    def EntryNodeOfLoop2(self, pHead):
        slow = pHead
        fast = pHead
        while fast != None and fast.next != None:
            fast = fast.next.next
            slow = slow.next
            if slow == fast:
                # 途中根据关系可以推出a=c，所以只需要slow head都向前走，相等的点就是入口
                while pHead != slow:
                    slow = slow.next
                    pHead = pHead.next
                return slow
        return None
```

```python
# 6. 链表中倒数最后k个节点
# description：输入一个长度为 n 的链表，设链表中的元素的值为 ai ，返回该链表中倒数k个节点。
'''
思考： method1:快慢指针，可以使用双指针，指针差为k，
    method2.可以使用栈，（当有倒序的时候就使用栈）但是这样两者时间复杂度都是n，但前者空间复杂度为1，后者为n
'''

class Solution:
    def FindKthToTail(self , pHead: ListNode, k: int) -> ListNode:
        # fast slow 
        fast = pHead
        slow = pHead
        # fast runs extra k steps
        for i in range(0, k):
            if not fast:
                return None
            fast = fast.next
        # fast, slow run simutaneously
        while fast:
            fast = fast.next
            slow = slow.next
        return slow
        
```

```python
# 8. 删除链表中重复节点：medium
# description：在一个排序的链表中，存在重复的结点，请删除该链表中重复的结点，重复的结点不保留，返回链表头指针。 
'''
思考：升序链表，重复节点都连在一起，可以容易找到重复节点，将相连的相同节点跳过，连接不相同的第一个节点
    小tip：链表前加上表头，方便可能情况删除第一个节点
    遍历链表，比较相邻节点，遇到两个相邻节点相同，新开内循环，将相同的一段都过去
    返回时去掉表头
'''
class Solution:
    def deleteDuplication(self, pHead: ListNode) -> ListNode:
        if pHead == None:
            return None
        res = ListNode(0)
        res.next = pHead
        cur = res
        while cur.next and cur.next.next:
            if cur.next.val == cur.next.next.val:
                tmp = cur.next.val
                # 所有相同跳过
                while cur.next != None and cur.next.val == tmp:
                    cur.next = cur.next.next
            else:
                cur = cur.next
            return res.next
                    
```

```python
# 9. 删除链表的节点：easy
# description：给定单向链表的头指针和一个要删除的节点的值，定义一个函数删除该节点。返回删除后的链表的头节点。
'''
思考：删除链表元素的做法，先断掉它前一个节点指向它的指针，然后指向它后面一个节点，就越过了该节点
pre.next = cur.next
'''
class ListNode:
    def __init__(self, x):
        self.val = x
        self.next = None
        
class Solution:
    def deleteNode(self , head: ListNode, val: int) -> ListNode:
        # 加入头节点
        res = ListNode(0)
        res.next = head
        # pre & cur
        pre = res
        cur = head
        # 遍历链表
        while cur:
            if cur.val = val:
                pre.next = cur.next
                break
            # 没有到目标点，就都向前走一步
            pre = cur
            cur = cur.next
        # 返回去掉后的头节点,res的是不变的
        return res.next
```
